## Example: Sum of Bernoulli Random Variables
Consider a random sample $(X_1, X_2, \ldots, X_n)$ from a Bernoulli distribution with parameter \(p\). The sum of these variables, $(Y = \sum_{i=1}^n X_i)$, follows a binomial distribution with parameters \(n\) and \(p\).

### Moment Generating Functions (MGFs) 
Moment Generating Function (MGF): 
 - Denoted by $(M_X(t))$, it is defined as $(E[e^{tX}])$.
 - MGFs uniquely determine the distribution of a random variable.

#### Example: MGF of Bernoulli Distribution

````r
# MGF of Bernoulli distribution
bernoulli_mgf <- function(t, p) {
  (1 - p) + p * exp(t)
}

# Calculate MGF for t = 1
t <- 1
mgf_value <- bernoulli_mgf(t, p)
cat(sprintf("MGF of Bernoulli distribution at t = %d: %.3f\n", t, mgf_value))
````

### Sum of IID Random Variables
For a random sample $(X_1, X_2, ..., X_n)$ from any distribution with MGF $(M_X(t))$, the MGF of the sum

$$(Y = \sum_{i=1}^n X_i)$$

is

 
$$(M_Y(t) = [M_X(t)]^n).$$

### Example: Sum of IID Exponential Random Variables
Consider a random sample from the exponential distribution with rate ${\lambda}$.

#### Example in R: Sum of Exponential Random Variables

````r
# Parameters
lambda <- 1
n <- 10

# Generate a random sample from Exponential distribution
set.seed(123)
sample_data <- rexp(n, rate = lambda)

# Sum of the sample
sum_sample <- sum(sample_data)
cat(sprintf("Sum of Exponential sample: %.3f\n", sum_sample))
````

## Gamma Distribution
The sum of $(n) (iid)$ exponential random variables with rate $(\lambda)$ follows a gamma distribution with shape parameter $(n)$ and rate parameter $(\lambda)$.

#### Example in R: Gamma Distribution

````r
# Parameters
alpha <- n
beta <- lambda

# Generate a random sample from Gamma distribution
set.seed(123)
gamma_sample <- rgamma(1000, shape = alpha, rate = beta)

# Plot the Gamma distribution
library(ggplot2)
ggplot(data.frame(x = gamma_sample), aes(x = x)) +
  geom_histogram(aes(y = after_stat(density)), bins = 30, fill = "blue", alpha = 0.5) +
  geom_density(color = "red", linewidth = 1) +
  labs(title = "Gamma Distribution",
       x = "Value",
       y = "Density")
````
