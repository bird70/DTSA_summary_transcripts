# Mod4/L3: Statistical Inference: T-Distribution and Confidence Intervals

## 1. Large Sample Theory

### Prerequisites

Consider a random sample of size n (IID) from any distribution with:
- Mean μ 
- Finite variance σ²
- Sample size n > 30-40 generally considered "large"
- Larger samples recommended for critical applications
- Visual inspection of histograms advised for asymptotic normality assumption

### Central Limit Theorem (CLT)
- For large n, X̄ is asymptotically/approximately normal
- Properties:
  - Mean = μ (original distribution mean)
  - Variance = σ²/n (original variance/n))
  - Standardized form: (X̄ - μ)/(σ/√n) → N(0,1)

## 2. Sample Variance and Large Sample CI

### Sample Variance Properties
- s² is unbiased estimator of σ²
- s² converges to σ² in probability
- Can substitute s for σ in large samples

### Large Sample CI Formula
(1-α)100% CI:
```
 X̄ ± z_(α/2) * (s/√n)
```

## 3. Small Sample Theory

### Requirements
- Must know underlying distribution (typically normal)
- Both μ and σ² unknown
- Sample size too small for CLT

### T-Distribution Method
- Use t-distribution with (n-1) degrees of freedom
- CI Formula: 
```
X̄ ± t_(α/2,n-1) * (s/√n)
```

## 4. Practical Example

### Glucose Monitoring System Study
- Sample size: n = 20
- Sample mean: X̄ = 187 days
- Sample variance: s² = 16.2 days
- Confidence level: 95% (α = 0.05)

### Results
- Critical value: t_(0.975,19) = 2.09
- 95% CI: [185.11, 188.88] days

## 5. Key Takeaways

1. Large samples (n > 30) use z-distribution
2. Small samples require t-distribution
3. Normal distribution assumption critical for small samples
4. CI interpretation: range of plausible parameter values
