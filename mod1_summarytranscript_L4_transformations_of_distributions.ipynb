# Mod1/L4 Transformations of Distributions

## Introduction
In this lesson, we explore transformations of distributions. Given a random variable \(X\), we aim to determine the distribution of a new random variable \(Y\) obtained by applying a function to \(X\).

## Discrete Case

### Example: Binomial Distribution
The binomial distribution arises from a sequence of \(n\) independent trials, each with a success probability \(p\). The random variable \(X\) counts the number of successes in \(n\) trials.

#### Probability Mass Function (PMF)
For $(X \sim \text{Binomial}(n, p))$:
$$[ P(X = x) = \binom{n}{x} p^x (1-p)^{n-x} ]$$

### Transformation Example
Suppose $$(Y = n - X)$$ We want to find the distribution of \(Y\).

#### Transformation Steps
1. **PMF for \(X\)**:
   $$[ P(X = x) = \binom{n}{x} p^x (1-p)^{n-x} ]$$
2. **Express \(Y\) in terms of \(X\)**:
   $$[ Y = n - X ]$$
3. **Find the PMF for \(Y\)**:
   $$[ P(Y = y) = P(X = n - y) = \binom{n}{n-y} p^{n-y} (1-p)^y ]$$
4. **Simplify**:
   $$[ P(Y = y) = \binom{n}{y} (1-p)^y p^{n-y} ]$$
   Thus, $$(Y \sim \text{Binomial}(n, 1-p)).$$

## Continuous Case

### Cumulative Distribution Function (CDF)
The CDF $$(F_X(x))$$ of a continuous random variable \(X\) is:
$$[ F_X(x) = P(X \leq x) ]$$

### Transformation Example
Suppose $$(Y = g(X))$$ where \(g\) is a continuous, differentiable, and invertible function.

#### Steps for Transformation
1. **CDF for \(Y\)**:
   $$[ F_Y(y) = P(Y \leq y) = P(g(X) \leq y) ]
2. **Solve for \(X\)**:
   $$[ g(X) \leq y \implies X \leq g^{-1}(y) ]
   $$[ F_Y(y) = F_X(g^{-1}(y)) ]
3. **PDF for \(Y\)**:
   $$[ f_Y(y) = \frac{d}{dy} F_Y(y) = f_X(g^{-1}(y)) \left| \frac{d}{dy} g^{-1}(y) \right| ]$$

### Example in R: Gamma Distribution
Suppose 
$$(X \sim \text{Gamma}(\alpha, \beta)) and (Y = cX).$$

#### Transformation Steps
1. **Function and Inverse**:
   $$[ Y = cX \implies X = \frac{Y}{c} ]$$
2. **PDF for \(Y\)**:
   $$[ f_Y(y) = f_X\left(\frac{y}{c}\right) \left| \frac{d}{dy} \left(\frac{y}{c}\right) \right| ]$$
   $$[ f_Y(y) = \frac{\beta^\alpha}{\Gamma(\alpha)} \left(\frac{y}{c}\right)^{\alpha-1} e^{-\beta \frac{y}{c}} \frac{1}{c} ]$$
   $$[ f_Y(y) = \frac{\beta^\alpha}{c^\alpha \Gamma(\alpha)} y^{\alpha-1} e^{-\frac{\beta}{c} y} ]$$
   Thus, $$(Y \sim \text{Gamma}(\alpha, \frac{\beta}{c})).$$

#### Example in R
```r
# Parameters
alpha <- 2
beta <- 1
c <- 2

# Generate a random sample from Gamma distribution
set.seed(123)
sample_data <- rgamma(1000, shape = alpha, rate = beta)

# Transform the sample
transformed_sample <- c * sample_data

# Plot the transformed Gamma distribution
library(ggplot2)
ggplot(data.frame(x = transformed_sample), aes(x = x)) +
  geom_histogram(aes(y = ..density..), bins = 30, fill = "blue", alpha = 0.5) +
  geom_density(color = "red", size = 1) +
  labs(title = "Transformed Gamma Distribution",
       x = "Value",
       y = "Density")

## Conclusion
Transformations of distributions allow us to understand how applying functions to random variables affects their distributions. By using the CDF and PDF, we can derive the distribution of the transformed variable. This is particularly useful in statistical analysis and data transformations.
