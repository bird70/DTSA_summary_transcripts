# Mod4/L3: T-Distribution and Confidence Intervals

## (1) Large Sample Theory

Consider a random sample of size n (IID) from any distribution with:
- Mean μ 
- Finite variance σ²

### Large Sample Properties
- Sample size typically n > 30 or 40
- For critical applications, larger samples recommended
- Visual inspection of histograms advised for asymptotic normality assumption

## Central Limit Theorem Application

For large samples:
- X̄ is asymptotically normal
- Mean = μ (original distribution mean)
- Variance = σ²/n (original variance/n)

### Key Point on Asymptotic Normality
The standardized quantity (X̄ - μ)/(σ/√n) converges in distribution to N(0,1)

## Using Sample Variance

When σ² is unknown:
- Sample variance s² is an unbiased estimator of σ²
- s² converges in probability to σ²
- For large samples, can substitute s for σ

### Large Sample Confidence Interval

The approximate (1-α)100% confidence interval for μ is:
```
X̄ ± z_(α/2) * (s/√n)
```

## (2) Small Sample Theory

### Requirements for Small Samples
- Must know underlying distribution
- Normal distribution case most common
- Both μ and σ² unknown

### T-Distribution Application

For normal samples with unknown σ²:
- Use t-distribution with (n-1) degrees of freedom
- Critical values from t-distribution tables/software
- Confidence interval: 
```
X̄ ± t_(α/2,n-1) * (s/√n)
```

## (3) Practical Example

Glucose monitoring system sensor study:
- n = 20 patients
- X̄ = 187 days (average lifetime)
- s² = 16.2 days (sample variance)
- α = 0.05 (95% confidence level)

### Calculation
Using R: `qt(0.975, 19)` ≈ 2.09

95% Confidence Interval: [185.11, 188.88] days
