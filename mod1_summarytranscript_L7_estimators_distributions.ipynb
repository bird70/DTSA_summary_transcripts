## Introduction
In this lesson, we explore the concept of estimators and their distributions. We aim to understand why certain statistics are used to estimate population parameters and how their sampling distributions behave.

## Why Use Samples?
To estimate the true average height of all adult males in France, we take a random sample and compute the sample mean. This common-sense approach allows us to make inferences about the population.

## Key Concepts

### 1. Statistics and Estimators
- **Statistic**: Any function of the sample data (e.g., sample mean, sample median).
- **Estimator**: A random variable used to estimate a population parameter before sampling.

### 2. Sampling Distribution
- The distribution of a statistic over many samples.
- Helps us understand the variability and reliability of the statistic.

## Notation
- **Random Sample**: A collection of n random variables drawn from a population.
- **Independent and Identically Distributed (iid)**: Assumes each sample is independent and follows the same distribution.

````r
# Generate a random sample from Gamma distribution
set.seed(123)
n <- 100
alpha <- 2
beta <- 1
sample_data <- rgamma(n, shape = alpha, rate = beta)

# Display sample data
head(sample_data)
````

## Estimators and Their Properties
1. **Unbiased Estimator**
An estimator is unbiased if its expected value equals the true parameter value.

#### Example in R: Sample Mean as an Unbiased Estimator

````r
# Sample mean
sample_mean <- mean(sample_data)

# Expected value of sample mean
expected_value <- alpha / beta
cat(sprintf("Sample Mean: %.3f, Expected Value: %.3f\n", sample_mean, expected_value))
````

2. **Variance of an Estimator** The variance of an estimator indicates its spread and reliability. A smaller variance means the estimator is more predictable.

#### Example in R: Variance of Sample Mean

````r
# Variance of sample mean
sample_variance <- var(sample_data)
sample_mean_variance <- sample_variance / n
cat(sprintf("Sample Variance: %.3f, Variance of Sample Mean: %.3f\n", sample_variance, sample_mean_variance))
````

## Sampling Distribution of the Sample Mean
The sample mean has its own distribution, known as the sampling distribution. It has a mean equal to the population mean and a variance equal to the population variance divided by the sample size.

### Example in R: Sampling Distribution of the Sample Mean

````r
# Simulate sampling distribution of the sample mean
set.seed(123)
num_simulations <- 1000
n<-100
sample_means <- replicate(num_simulations, mean(rgamma(n, shape = alpha, rate = beta)))

# Plot sampling distribution
library(ggplot2)
ggplot(data.frame(sample_means), aes(x = sample_means)) +
  geom_histogram(aes(y = ..density..), bins = 30, fill = "blue", alpha = 0.5) +
  geom_density(color = "red", size = 1) +
  labs(title = "Sampling Distribution of the Sample Mean",
       x = "Sample Mean",
       y = "Density")
````

## Conclusion
Understanding the properties of estimators and their distributions is crucial for making reliable inferences about population parameters. The sample mean is a commonly used estimator due to its unbiased nature and predictable variance.
